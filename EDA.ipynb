```python
# Group by YEAR and MONTH and count the number of occurrences
monthly_crime_counts = df_cleaned.groupby(['YEAR', 'MONTH']).size().reset_index(name='CRIME_COUNT')

# Sort by CRIME_COUNT to find the month with the highest crime count
monthly_crime_counts_high_low = monthly_crime_counts.sort_values(by='CRIME_COUNT', ascending=False)

# Display the results
monthly_crime_counts_high_low.head(10)
```

```python
import matplotlib.pyplot as plt

# Plotting the monthly crime counts
plt.figure(figsize=(14, 8))
for year in monthly_crime_counts['YEAR'].unique():
    subset = monthly_crime_counts[monthly_crime_counts['YEAR'] == year]
    plt.plot(subset['MONTH'], subset['CRIME_COUNT'], marker='o', label=f'Year {year}')

plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.title('Monthly Crime Counts')
plt.legend()
plt.grid(True)
plt.show()
```

```python
crime_by_month = df.groupby(['MONTH', 'TYPE']).size().unstack().fillna(0)

crime_by_month.plot(kind='bar',stacked=True, figsize=(12, 6))
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.title('Crime Distribution by Month')
plt.legend(title='Crime Type', bbox_to_anchor=(1.03, 1), loc='upper left')
plt.show()

```
